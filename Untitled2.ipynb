```python
import pandas as pd
import io
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import pandas as pd
import io
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns


# Load dataset
filename = next(iter(uploaded))  # Gets the uploaded file name
df = pd.read_csv(io.BytesIO(uploaded[filename]), header=None, encoding='latin-1') # Changed encoding to 'latin-1'

# Rename columns for clarity before accessing them
df = df.rename(columns={0: 'sentiment', 1: 'text'})

# Print unique sentiment values to check for case sensitivity and typos
print(df['sentiment'].unique())

# Keep only relevant columns
df = df[['text', 'sentiment']]

# Filter to only positive, negative, neutral - Adjust values if needed based on the unique values printed above
df = df[df['sentiment'].isin(['positive', 'negative', 'neutral'])] # Changed to lowercase

# Normalize sentiment labels
df['sentiment'] = df['sentiment'].str.lower()  # to 'positive', 'negative', 'neutral'

# Drop missing values
df = df.dropna()

# Check if the DataFrame is empty after filtering
if df.empty:
    print("DataFrame is empty after filtering. Check your sentiment values and filtering conditions.")
else:
    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(df['text'], df['sentiment'], test_size=0.2, random_state=42)

    # Create a TF-IDF vectorizer to convert text into numerical features
    vectorizer = TfidfVectorizer()
    X_train_vec = vectorizer.fit_transform(X_train)
    X_test_vec = vectorizer.transform(X_test)

    # Train a Multinomial Naive Bayes model
    model = MultinomialNB()
    model.fit(X_train_vec, y_train)

    # Predict on test data
    y_pred = model.predict(X_test_vec)

    # Show report
    print("\nClassification Report:\n")
    print(classification_report(y_test, y_pred))

    # Plot sentiment distribution
    plt.figure(figsize=(8, 6))
    sns.countplot(x=y_pred, hue=y_pred, palette='Set2', legend=False)

    plt.title('Predicted Sentiment Distribution')
    plt.xlabel('Sentiment')
    plt.ylabel('Number of Tweets')
    plt.show()
```
